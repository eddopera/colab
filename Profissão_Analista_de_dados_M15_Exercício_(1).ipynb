## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.boxplot(x=diamonds["price"])
plt.title("Distribuição dos preços dos diamantes (outliers)")
plt.xlabel("Preço")
plt.show()
```

**Insight do gráfico 1**: ...  O gráfico boxplot mostra que existem diversos outliers nos preços dos diamantes, principalmente na faixa de preços mais altos, indicando que alguns diamantes são vendidos por valores muito superiores à maioria.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.scatterplot(x="carat", y="price", hue="clarity", data=diamonds, palette="tab10")
plt.title("Correlação entre preço e peso (carat) por transparência (clarity)")
plt.xlabel("Peso (carat)")
plt.ylabel("Preço")
plt.legend(title="Transparência (clarity)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Insight do gráfico 2**: ... Existe uma forte correlação positiva entre o peso do diamante (carat) e seu preço, independentemente da transparência (clarity). Diamantes mais pesados tendem a ser mais caros. Além disso, observa-se que para um mesmo peso, diferentes níveis de transparência influenciam o preço, mas o peso é o fator mais determinante.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

import seaborn as sns
import matplotlib.pyplot as plt

cars = sns.load_dataset("mpg")

plt.figure(figsize=(10, 6))
sns.kdeplot(cars["weight"], fill=True)
plt.title("Distribuição contínua aproximada do peso dos carros")
plt.xlabel("Peso")
plt.ylabel("Densidade")
plt.show()
```

**Insight do gráfico 3**: ... A distribuição do peso dos carros é assimétrica, com a maioria dos veículos concentrada em uma faixa de peso intermediário e uma cauda mais longa para carros mais pesados. Isso indica que carros leves são menos comuns e há poucos veículos extremamente pesados na amostra.
