## 1\. Preço do diamante por tipo de corte

````
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
````

````
# gráfico do exercício 1

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(8, 6))
sns.boxplot(x="cut", y="price", data=diamonds)
plt.title("Preço do diamante por tipo de corte")
plt.xlabel("Tipo de corte")
plt.ylabel("Preço")
plt.show()
````

## 2\. Número de passageiros em dezembro por ano

````
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
````

````
# gráfico do exercício 2

import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

dezembro = flights[flights['month'] == 'December']

plt.figure(figsize=(8, 6))
plt.plot(dezembro['year'], dezembro['passengers'], marker='o')
plt.title("Número de passageiros em dezembro por ano")
plt.xlabel("Ano")
plt.ylabel("Número de passageiros")
plt.grid(True)
plt.show()
````

**Insight do gráfico 2**: ...

**negrito**---O número de passageiros em dezembro aumentou consistentemente ao longo dos anos, indicando crescimento do setor aéreo nesse período do ano.

## 3\. Numero de passageiros por mês entre 1949 e 1959

````
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

plt.figure(figsize=(12, 6))
sns.lineplot(x="year", y="passengers", hue="month", data=flights, marker="o")
plt.title("Número de passageiros por mês entre 1949 e 1959")
plt.xlabel("Ano")
plt.ylabel("Número de passageiros")
plt.legend(title="Mês", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````
